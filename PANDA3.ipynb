```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('D:\\excler_material\\files\\mtcars.csv')
```

```python
data
```

```python
plt.scatter(data.mpg,data.qsec)
plt.xlabel('mpeg')
plt.ylabel('qsec')
plt.title('scatter plot')
```

```python
plt.hist(data['mpg'],facecolor='peru',edgecolor='blue',bins=4,fill=False,hatch='\\')
```

```python
plt.boxplot(data[['mpg','qsec','gear','cyl']])
plt.title('Box Plot')
plt.ylabel('y-axis')
plt.xlabel('X-axis')
```

```python
plt.violinplot(data['mpg'])
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(y='tip',data=tips,jitter=True)
```

```python
sns.stripplot(x='tip',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue ='smoker')
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='time')
```

```python
sns.boxplot(x='day',y='tip',data= tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips)
```
